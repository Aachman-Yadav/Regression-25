# Visualization Feature Realtionship with Target Variable

**Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Using the dataset which i scaled in the previous notebook**

```python
df = pd.read_csv("wines_SPA_scaled_dataset.csv")
df.head()
```

**Assuming the target in this case is price, so we will follow the analysis for the price column.**

**As we can see Winery are have many unique values so we are going to remove it**

```python
df['winery'].nunique()
```

```python
df.shape
```

```python
df = df.drop(['winery'], axis = "columns")
```

```python
df.head()
```

**As we can see Wine column also has many unqiue so we are going to remove it as well**

```python
df['wine'].nunique()
```

```python
df = df.drop(['wine'], axis = "columns")
df.head()
```

**Type column has only 20 unique values so we are not going to remove it**

```python
df['type'].nunique()
```

```python
df['region'].nunique()
```

```python
df = df.drop(['region'], axis="columns")
df.head()
```

**Now for better visualization we are going to Encode type column**

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
```

```python
df['type_2'] = label.fit_transform(df['type'])
df.head()
```

```python
df = df.drop(['type'], axis="columns")
df.head()
```

**Making a heatmap to know the corelation between every columns with each other**

```python
plt.figure(figsize=(12, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```

**As we can see there no corelation higher than 75% so we can not decide perfect target and other columns which does not act as target**

**Making a Scatterplot To know if any other relation forms**

```python
def scatter(df , target):
  for i in df.columns:
    if i == target:
      continue
    plt.figure(figsize=(7, 5))
    sns.scatterplot(x=df[i], y=df[target])
    plt.title(f"{i} vs. Price")
    plt.show()
```

```python
scatter(df, 'price')
```

**There is not good corelation between any of the columns :(**
