# Using different ML models to get the maximum accuracy

**Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot
import seaborn as sns
```

**Importing the final dataset that i made in the last notebook**

```python
df = pd.read_csv("wines_SPA_final.csv")
df.head()
```

**This time assunming the target column to be the type of the wine**

```python
x = df.drop(['type_2'], axis="columns")
y = df['type_2']
```

**Splitting the dataset from the target**

```python
x.head()
```

```python
y.head()
```

**Using train_test split to split the dataset**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x , y, test_size=0.2, random_state=30)
```

```python
x_train.shape
```

```python
x_test.shape
```

**Firstly trying Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)
```

**Accuracy of Logistic Regression -> 0.5551020408163265**

```python
model.score(x_test, y_test)
```

**Now using Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

**Accuracy of Decision Tree -> 0.46938775510204084**

```python
model.score(x_test, y_test)
```

**Now using Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(n_estimators=100)
forest.fit(x_train, y_train)
```

**Accuracy of Random Forest -> 0.5714285714285714**

```python
forest.score(x_test, y_test)
```

**Now using Gradient Boosting Classifier**

```python
from sklearn.ensemble import GradientBoostingClassifier

gradient = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1)
gradient.fit(x_train, y_train)
```

**Accuracy of Gradient Boosting Classifier -> 0.5306122448979592**

```python
gradient.score(x_test, y_test)
```

**Now using Vector Machine**

```python
from sklearn.svm import SVC

vector = SVC(kernel='rbf', C=1.0)
vector.fit(x_train, y_train)
```

**Accuracy of Vector Machine 0.5755102040816327**

```python
vector.score(x_test, y_test)
```

**Now using KNN Classifier**

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
```

**Accuracy of KNN -> 0.49387755102040815**

```python
knn.score(x_test, y_test)
```

**After using various models the highest accuracy came of Support Vector Machine which was -> 0.5755102040816327**
